```python
df["Total_cases"] = df["Deaths"] + df["Recovered"] + df["Active"]
```

```python
df["Death_Rate"] = (df["Deaths"] / df["Total_cases"]) * 100
```

```python
low_death_rate_threshold = 2.0  


if "New_Cases" not in df.columns:
    df["New_Cases"] = df["Total_cases"].diff().fillna(0)  # Calculate new cases


low_death_rate_countries = df[df["Death_Rate"] < low_death_rate_threshold]


increasing_cases = df[df["New_Cases"] > 0]


result = increasing_cases[increasing_cases["Country/Region"].isin(low_death_rate_countries["Country/Region"])]

inplace=True
print(result[["Country/Region", "Total_cases", "New_Cases", "Death_Rate"]])
```
